```python
from dotenv import load_dotenv
from groq import Groq
import os
load_dotenv()

client = Groq(
    api_key=os.environ.get("GROQ_API")
)
```

```python
# com limite de formato e input
entrada = input("Digite uma palavra para criar uma caça palavras com palavras semelhantes")
prompt_com_limite = f"""Retorne 10 palavras que são extremamente relacionadas com a palavra:{entrada}

você irá retornar essas 10 palavras em um caça palavras, em espécie de matriz. 


vou te dar um exemplo do padrão que eu desejo:

imagine 2 palavras: gato e rato

T T A Y O P G
A A S R Q G A
R A T O A E T
E S R Q D G O

No caça palavras acima, eu tenho a palavra Rato e Gato. Faça o mesmo com as 10 palavras relacionadas a palavra informada

Responda APENAS com a matriz do caça palavras, sem explicações.

"""

response_detalhado = client.chat.completions.create(

model = "openai/gpt-oss-120b",
messages=[
    {"role":"system", "content":"você é um especialista criar caça palavras"}
    , {"role":"user", "content":prompt_com_limite}
],
    temperature=1
)

print(response_detalhado.choices[0].message.content)
```
