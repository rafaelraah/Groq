```python
from dotenv import load_dotenv
from groq import Groq
import os
load_dotenv()

client = Groq(
    api_key=os.environ.get("GROQ_API")
)
```

```python
system_prompt = """Você é um nutricionista virtual adaptativo. Siga estas condições:

SE a pergunta for BÁSICA:
- Explique de forma simples
- Use exemplos do dia a dia
- Evite termos técnicos

SE a pergunta for AVANÇADA:
- Use linguagem científica
- Cite termos nutricionais
- Sugira leituras adicionais

SE a pergunta for AMBÍGUA:
- Peça esclarescimentos
- Ofereça opções de interpretação

Me informar no final da resposta se você  considerou a pergunta básica, avançada ou ambígua, para dar a sua resposta


"""

user_prompt = """ O que é proteína? """

response = client.chat.completions.create(

model = "openai/gpt-oss-120b",
messages=[
    {"role":"system", "content":system_prompt}
    , {"role":"user", "content":user_prompt}
],
    temperature=0.7,
    ##max_completion_tokens=400
    
)

print("=== Exemplo 1: Básico ===")
print(response.choices[0].message.content)
```

### Adaptações ao Nível do Usuário

```python
system_prompt = """Você é o coach de saúde que adapta explicações conforme o nível:

INICIANTE (palavras: "novo", "começando", "não sei ainda"):
- explicações muito simples
- exemplos cotidianos
- passo a passo claro

INTERMEDIARIO (palavras: "já sei", "tenho alguma experiência"):
- Respostas diretas
- Pequenos detalhes técnicos
- Sugestões práticas

AVANÇADO (palavras: "experiente", "profissional", "avançado"):
- Respostas técnicas
- citações científicas
- estratégias avançadas

No final do texto, me fale qual você considerou para dar a resposta:
iniciante, intermediario ou avançado, e vai me explicar o motivo de ter tomado essa decisão

"""

user_prompt_iniciante = """ Sou novo em dieta, como começo a comer melhor? """
user_prompt_avancado = """ Sou nutricionista, explique estratégias de periodização alimentar """

response_iniciante = client.chat.completions.create(

model = "openai/gpt-oss-120b",
messages=[
    {"role":"system", "content":system_prompt}
    , {"role":"user", "content":user_prompt_iniciante}
],
    temperature=0.7,
    ##max_completion_tokens=400
    
)

response_avancado = client.chat.completions.create(

model = "openai/gpt-oss-120b",
messages=[
    {"role":"system", "content":system_prompt}
    , {"role":"user", "content":user_prompt_avancado}
],
    temperature=0.7,
    ##max_completion_tokens=400
    
)

print("=== Exemplo 1: INICIANTE ===")
print(response_iniciante.choices[0].message.content)

print("\n\n----------------\n\n")

print("=== Exemplo 2: AVANÇADO ===")
print(response_avancado.choices[0].message.content)
```

### usando Fallback

```python
system_prompt = """

Você é um chatbot de nutrição. Regras:

SE a pergunta for sobre RECEITAS:
- Sugira 3 opções básicas

SE for sobre EXERCÍCIOS:
- Recomende treinos simples

SE for sobre SUPLEMENTOS:
- Explique benefícios e riscos

CASO CONTRÁRIO (fallback):
- Peça mais detalhes
- Sugira opções de tema
- Sempre responda com simpatia

No final da resposta, me fale em qual condição caiu

"""

user_prompt = """Vocês podem me ajudar a correr mais rápido na lua?"""

response = client.chat.completions.create(

model = "openai/gpt-oss-120b",
messages=[
    {"role":"system", "content":system_prompt}
    , {"role":"user", "content":user_prompt}
],
    temperature=0.7,
    ##max_completion_tokens=400
    
)

print("=== Exemplo: FALLBACK ===")
print(response.choices[0].message.content)
```
