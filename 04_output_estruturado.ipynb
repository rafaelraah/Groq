```python
from dotenv import load_dotenv
from groq import Groq
import os
load_dotenv()

client = Groq(
    api_key=os.environ.get("GROQ_API")
)
```

### Saída em formato de tabela

```python
system_prompt = """Você é um historiador especializado em civilizações antigas.
Sempre responda no formato de tabela:

| Civilização | Período | Contribuição Marcante | Região |
|-------------|---------|-----------------------|--------|

"""

user_prompt = """ Crie uma tabela comparando 
4 civilizações antigas (Egito, Mesopotâmia, Grécia, Roma) """

response = client.chat.completions.create(

model = "openai/gpt-oss-120b",
messages=[
    {"role":"system", "content":system_prompt}
    , {"role":"user", "content":user_prompt}
],
    temperature=0.7,
    max_completion_tokens=400
    
)

print(f"Resposta: {response.choices[0].message.content}\n")
```
