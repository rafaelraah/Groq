```python
from dotenv import load_dotenv
from groq import Groq
import os
load_dotenv()

client = Groq(
    api_key=os.environ.get("GROQ_API")
)
```

### Parâmetros importantes
- Temperature: controla a aleatoridade das respostas. Valores entre 0.0 e 1.0 são aceitos (Embora algumas implementações suportem valores até 2, o padrão recomendado para o Groq é manter-se no intervalo de 0 à 1 para garantir a coerência.)
- max_completion_tokens: define o número máximo de tokens que podem ser gerados pelo modelo. Esse parametro substitui o max_tokens em desuso.
- top_p: controla a diversidade via "nucleus sampling" - em torno de 0.5 significa considerar apenas os tokens que somam 50% de probabilidade

## Usando Temperature 0.0

```python
prompt = "Escreva uma frase poética sobre o anoitecer"

# Temperatura baixa (0.0) mais determinística
print("Temperatura 0.0 (determinística):")
response_low = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
        {"role": "system", "content":"Você é um poeta minimalista que descreve a natureza de forma simples e direta."},
        {"role": "user", "content": prompt}   
    ],
    temperature=0.0
)
print(f"Resposta: {response_low.choices[0].message.content}\n")
```

## Usando Temperature 0.7

```python
prompt = "Escreva uma frase poética sobre o anoitecer"

# Temperatura média (0.7) mais equilibrada
print("Temperatura 0.7 (equilibrada):")
response_low = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
        {"role": "system", "content":"Você é um poeta clássico que mistura metáforas e imagens da natureza."},
        {"role": "user", "content": prompt}   
    ],
    temperature=0.7
)
print(f"Resposta: {response_low.choices[0].message.content}\n")
```

## Usando top_p

```python
prompt = "Complete a frase: 'A Inteligência artificial no futuro será...'"

# Top_p baixo (0.1) - Mais focado
print("Top_p 0.1 (Focado):")
response_low = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
        {"role": "system", "content":"Você é um futurista que faz previsões sérias e objetivas sobre tecnologia."},
        {"role": "user", "content": prompt}   
    ],
    temperature=1.0, #fixo para comparar só o efeito do top_p
    top_p=0.1
)
print(f"Resposta: {response_low.choices[0].message.content}\n")
```
