### Sumarização

```python
from dotenv import load_dotenv
from groq import Groq
import os
load_dotenv()

client = Groq(
    api_key=os.environ.get("GROQ_API")
)
```

### Sumarização Básica

```python
system_prompt = """

Você é um editor técnico especializado em jogos independentes.
Crie um resumo objetivo com aproximadamente 1/3 do tamanho do texto original, preservando os pontos centrais e o tom informativo.

"""

user_prompt = """A cena dos jogos independentes em 2026 está mais criativa, 
caótica e divertida do que nunca. Pequenos estúdios e até desenvolvedores solo 
estão conseguindo criar experiências que muitas vezes chamam mais atenção do que
grandes produções milionárias. Com ferramentas acessíveis, inteligência artificial
auxiliando no desenvolvimento e motores gráficos cada vez mais poderosos, 
criar um jogo deixou de ser algo exclusivo de grandes empresas.

Hoje é comum encontrar jogos independentes com histórias emocionantes,
gráficos estilizados e ideias completamente malucas que dificilmente 
seriam aprovadas por grandes publishers. Existem jogos sobre administrar
uma cafeteria para gatos mágicos, sobreviver em um planeta feito de gelatina 
ou até simuladores absurdamente específicos, como organizar cabos de computador
ou trabalhar como entregador em cidades futuristas.

A comunidade indie também cresceu muito graças às plataformas digitais e 
às redes sociais. Um pequeno vídeo viral pode transformar um projeto desconhecido 
em um fenômeno mundial da noite para o dia. Muitos jogadores estão cansados de jogos 
genéricos e começaram a procurar experiências mais autênticas, criativas e pessoais, 
algo que os desenvolvedores independentes conseguem entregar muito bem.

Outro ponto interessante em 2026 é a mistura entre inteligência artificial e 
criatividade humana. Muitos criadores usam IA para gerar ideias, sons, diálogos
e protótipos rapidamente, permitindo que equipes pequenas façam jogos enormes em 
menos tempo. Mesmo assim, o toque humano continua sendo o diferencial que faz um 
jogo indie ter personalidade e conquistar fãs apaixonados.

A verdade é que os jogos independentes viraram uma espécie de “laboratório da indústria”.
É lá que surgem as ideias mais ousadas, os estilos visuais mais diferentes e as mecânicas
que mais tarde acabam influenciando toda a indústria dos games. Enquanto grandes empresas
jogam no seguro, os indies continuam experimentando, inovando e surpreendendo jogadores 
no mundo inteiro."""

response = client.chat.completions.create(

model = "openai/gpt-oss-120b",
messages=[
    {"role":"system", "content":system_prompt}
    , {"role":"user", "content":user_prompt}
],
    temperature=0.5,
    ##max_completion_tokens=400
    
)

print("=== Sumarização ===")
print(response.choices[0].message.content)
```

```python
texto = """Os jogos independentes, conhecidos como jogos “indie”, representam um dos segmentos mais criativos e inovadores da indústria dos games. Diferente das grandes produções financiadas por empresas bilionárias, os jogos independentes normalmente são desenvolvidos por pequenos estúdios ou até por uma única pessoa. Esse modelo permite maior liberdade criativa, dando espaço para ideias diferentes, estilos artísticos únicos e experiências inovadoras.
O crescimento desse segmento afeta diversos setores da economia e da tecnologia. A indústria de software é diretamente impactada, já que o desenvolvimento de jogos movimenta áreas como programação, inteligência artificial, computação gráfica e desenvolvimento de motores gráficos. O setor artístico também é fortemente beneficiado, envolvendo profissionais de design, animação, música, roteiro e produção audiovisual. Além disso, plataformas digitais de distribuição, criadores de conteúdo, streamers e campeonatos online também se fortalecem com o sucesso dos jogos independentes.
Outro setor impactado é o da educação. Muitos jogos independentes são utilizados como ferramentas de aprendizagem, estimulando criatividade, raciocínio lógico, resolução de problemas e até o aprendizado de idiomas. Existem também jogos voltados para saúde mental, treinamento profissional e simulações educacionais, mostrando que os games vão além do entretenimento.
Entre os principais benefícios dos jogos independentes está a inovação. Como pequenos desenvolvedores possuem mais liberdade, eles conseguem experimentar mecânicas, narrativas e estilos visuais que grandes empresas normalmente evitam por questões financeiras. Isso faz com que o mercado indie funcione como um laboratório de ideias para toda a indústria dos games. Outro benefício importante é a democratização do desenvolvimento de jogos. Com motores gráficos acessíveis e plataformas digitais modernas, qualquer pessoa com conhecimento e dedicação pode criar e publicar seu próprio jogo.
Entretanto, esse segmento também enfrenta desafios importantes. O primeiro deles é o financiamento. Muitos desenvolvedores independentes trabalham com baixo orçamento, precisando acumular várias funções ao mesmo tempo. A divulgação também é um problema, já que milhares de jogos são lançados todos os anos, tornando difícil ganhar visibilidade no mercado. Além disso, existe a pressão psicológica e financeira sobre equipes pequenas, que muitas vezes dependem do sucesso de um único projeto para continuar existindo.
Mesmo com esses desafios, a cena de jogos independentes continua crescendo e influenciando fortemente a indústria mundial. Muitos dos jogos mais criativos e memoráveis dos últimos anos surgiram justamente desse segmento, mostrando que boas ideias e paixão pelo desenvolvimento podem competir com grandes produções milionárias."""

# prompt pedindo JSON estruturado
user_prompt = f""" 
Resuma o seguinte texto em JSON
O Json deve conter as chaves:
- "setores" (lista com os setores impactados)
- "beneficios" (lista com os principais benefícios citados)
- "desafios" (lista com os riscos ou preocupações levantadas)

texto:
{texto}

"""

response = client.chat.completions.create(

model = "openai/gpt-oss-120b",
messages=[
     {"role":"user", "content":user_prompt}
],
    temperature=0.5,
    ##max_completion_tokens=400
    
)

print("=== Sumarização em formato json ===")
print(response.choices[0].message.content)
```
